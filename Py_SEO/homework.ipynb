## 作业2 sitemap模板制作

```python
# format方法
loc = 'www.baidu.com'
lastmod = '2020-09-11'
changefreq = 'daily'
prioritys = '1.0'
sitemap="""<?xml version="1.0" encoding="UTF-8"?>
<urlset>
    <url>
        <loc>{}</loc>
        <lastmod>{}</lastmod>
        <changefreq>{}</changefreq>
        <priority>{}</priority>
    </url>
</urlset>
""".format(loc, lastmod, changefreq, prioritys)
print(sitemap)
```

```python
# f-string放啊
loc = 'www.baidu.com'
lastmod = '2020-09-11'
changefreq = 'daily'
prioritys = '1.0'
sitemap=f"""<?xml version="1.0" encoding="UTF-8"?>
<urlset>
    <url>
        <loc>{loc}</loc>
        <lastmod>{lastmod}</lastmod>
        <changefreq>{changefreq}</changefreq>
        <priority>{prioritys}</priority>
    </url>
</urlset>
"""
print(sitemap)
```

```python
# %s 方法
loc = 'www.baidu.com'
lastmod = '2020-09-11'
changefreq = 'daily'
prioritys = '1.0'
sitemap=f"""<?xml version="1.0" encoding="UTF-8"?>
<urlset>
    <url>
        <loc>%s</loc>
        <lastmod>%s</lastmod>
        <changefreq>%s</changefreq>
        <priority>%s</priority>
    </url>
</urlset>
""" % (loc, lastmod, changefreq, prioritys)
print(sitemap)
```

# 第四天 列表

## 作业一 整理方法

```python
l1 = [3,2,1]
l2 = [4,5,6]
# 添加元素
l1.append(10)
print('l1添加10:',l1)
# 列表追加
l1.extend(l2)
print('l1追加l2:',l1)
# 列表相连
l3=l1+l2
print('l1追加l2:',l3)
# 删除元素值等于2的元素
l1.remove(2)
print('l1删除2:',l1)
# 推出最后一个元素
l1.pop()
print('l1推出最后一位元素:',l1)
# 当不用切片赋值时 赋值的内存地址 类似于指针
l3=l1
print('l1赋值给l3:',l3)
l3[0]=9
print('l1和l3:',l1,l3)
# 排血
l3.sort() # 升序
print('l3升序:',l3)
l3.sort(reverse=True) # 降序
print('l3降序:',l3)
sort1=sorted(l3)
print('l3升序:',sort1)
# 列表反转 返回迭代器
iter1=reversed(l3)
print('l3反转:')
for x in iter1:
    print(x)
```

## 作业二 九九乘法

```python
i = 9
while i > 0:
    j = i
    while j > 0:
        print(f'{i}*{j}={i*j}',end='\t')
        j -= 1
    print()
    i -= 1
```
